7.Vacuum Cleaner Agent Program

```python
# Vacuum Cleaner Agent Program

room_A = "Dirty"
room_B = "Dirty"
vacuum_location = "A"

print("Initial State:")
print("Room A:", room_A)
print("Room B:", room_B)
print("Vacuum Location:", vacuum_location)

# If vacuum is in Room A
if vacuum_location == "A":
    if room_A == "Dirty":
        print("\nCleaning Room A...")
        room_A = "Clean"
    print("Moving to Room B")
    vacuum_location = "B"

# If vacuum is in Room B
if vacuum_location == "B":
    if room_B == "Dirty":
        print("\nCleaning Room B...")
        room_B = "Clean"

print("\nFinal State:")
print("Room A:", room_A)
print("Room B:", room_B)
print("Vacuum Cleaner Task Completed ðŸ§¹")
```

11.Sum of First N Numbers

```python
n = int(input("Enter a number: "))
total = 0
for i in range(1, n + 1):
    total += i
print("Sum:", total)
print("Done")
```

12.Count Digits in a Number

```python
num = int(input("Enter a number: "))
count = 0
while num > 0:
    count += 1
    num //= 10
print("Digits:", count)
print("End")
```

13.Palindrome

```python
word = input("Enter a word: ")
rev = word[::-1]
if word == rev:
    print("Palindrome")
else:
    print("Not Palindrome")
print("Done")
```

14.Count Vowels

```python
text = input("Enter text: ")
vowels = "aeiouAEIOU"
count = 0
for ch in text:
    if ch in vowels:
        count += 1
print("Vowels:", count)
print("Finished")
```

15.Check Prime Number

```python
num = int(input("Enter a number: "))
flag = True

for i in range(2, num):
    if num % i == 0:
        flag = False
        break

if flag and num > 1:
    print("Prime Number")
else:
    print("Not a Prime Number")
```
